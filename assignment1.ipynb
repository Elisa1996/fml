```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df1 = pd.read_csv('Assignment 1 - steps data.csv', delimiter=";") # adding the steps data set
df1.head() # showing the first 5 to check if it works
```

```python
df2 = pd.read_csv('Assignment 1 - survey data.csv') # adding the survey data set
df2.head() # showing the first 5 to check if it works
```

```python
df = pd.merge(df1, df2, on = 'id') # merging the two dataframes together
df.head()
```

```python
sns.distplot(df['weight'].dropna(), kde=False) # select the weight column
plt.title('Weight of participants') # giving the graph a title
plt.xlabel('Weight (kg)') # adding label to the axe
plt.show() # show the graph
```

```python
sns.distplot(df['weight'], hist=True, kde=True, # creating a histogram for weight
             bins=int(100/5), # concentrating participants
             hist_kws={'edgecolor':'green'}) # adding an edgecolor to show the difference between bars
plt.title('Histogram of weight')
plt.xlabel('Weight (kg)')
plt.ylabel('Participants')
```

```python
sns.catplot(x="gender", y="weight", data=df, kind="violin", split=True,
height=6); # sizing the plot
print("The woman plot looks like a violin while the male plot is more like a square")
print("On average men have a higher wight than woman")
```

```python
print("WEIGHT STATS")
median_weight = df["weight"].median()
print("The median weight is", median_weight, "kg")
min_weight = df["weight"].min()
print("The lighest person has a weight of", min_weight, "kg")
max_weight = df["weight"].max()
print("The heaviest person has a weight of", max_weight, "kg")
sum_weight = df["weight"].sum()
print("The weight of all the 900 participants together is", sum_weight, "kg")
```
